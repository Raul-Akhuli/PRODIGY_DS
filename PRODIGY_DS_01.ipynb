```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
ages = [23, 45, 56, 45, 34, 23, 34, 56, 67, 78, 45, 56, 23, 34, 45, 67, 78, 89, 90, 45]
genders = ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']


data = pd.DataFrame({'Age': ages, 'Gender': genders})
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribution of Ages in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=data)
plt.title('Distribution of Genders in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
