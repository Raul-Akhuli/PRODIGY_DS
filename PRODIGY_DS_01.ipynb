# Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

population per year

```python
import pandas as pd

df = pd.read_csv('data1.csv', skiprows=4)
```

```python
df = df[['Country Name'] + [str(year) for year in range(1960, 2023)]]
```

```python
df = df.dropna()
```

```python
for year in range(1960, 2023):
    df[str(year)] = pd.to_numeric(df[str(year)], errors='coerce')
```

```python
global_population = df.drop('Country Name', axis=1).sum()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(14, 6))
plt.bar(global_population.index, global_population.values / 1e9, color='skyblue')
plt.title('Total Global Population Over Time (19602022)')
plt.xlabel('Year')
plt.ylabel('Population (Billion)')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dg = pd.read_csv('data1.csv', skiprows=4)
```

```python
country_name = 'India'
country_data = df[df['Country Name'] == country_name]
```

```python
years = [str(year) for year in range(1960, 2023)]
population_by_year = country_data[years].T
population_by_year.columns = ['Population']
population_by_year.index = population_by_year.index.astype(int)
```

```python
population_by_year['Population'] = population_by_year['Population'] / 1000000000
```

```python
plt.figure(figsize=(14, 6))
plt.bar(population_by_year.index, population_by_year['Population'], color='skyblue')

plt.title(f'{country_name} Population Over Time (19602022)', color='Black')
plt.xlabel('Year', color='Black')
plt.ylabel('Population (Billion)', color='Black')
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.xticks(rotation=45, color='white')
plt.yticks(color='Black')

plt.tight_layout()
plt.show()
```

for a specific year top countries population

```python
dh = pd.read_csv('data1.csv', skiprows=4)
```

```python
dh = dh[['Country Name', '2022']]
```

```python
dh = dh.dropna(subset=['2022'])
```

```python
dh['2022'] = dh['2022'].astype(int)
```

```python
dh_sorted = dh.sort_values(by='2022', ascending=False)

dh_top10 = dh_sorted.head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(dh_top10['Country Name'], dh_top10['2022'], color='skyblue')
plt.title('Top 10 Most Populous Countries in 2022')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
